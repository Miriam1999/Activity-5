## Tarea
Considere los siguientes datos y diseñe un clasificador binario por regresión logística.

Mostrar el gráfico de la división del clasificador con los puntos de entrenamiento.

Abrir un nuevo notebook, y subirlo a moodle en el espacio habilitado.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy.optimize as opt
X=10*np.random.random((100, 2))
Y=(X[:,1]>X[:,0]**2)*1
plt.figure(figsize=(8,6))
plt.scatter(X[:,0],X[:,1],c=Y)
plt.show()
```

```python
def fun_log(z):
    return 1/(1+np.exp(-z))
def reg_log(B,X):
    Xa=np.append(np.ones((X.shape[0],1)),X,axis=1)
    return fun_log(Xa.dot(B))
def fun_obj(B,X,Y):
    Yhat=0.0099*reg_log(B,X)
    return np.mean(-Y*np.log(Yhat)-(1-Y)*np.log(1-Yhat))
B0=np.zeros((3,))
res_clas=opt.minimize(fun_obj,B0,args=(X,Y))
res_clas
```

```python
B=res_clas.x
B
```

```python
x=np.arange(2,11,0.05)
y=np.arange(2,11,0.05)
Xm,Ym=np.meshgrid(x,y)
m,n=np.shape(Xm)
Xmr=np.reshape(Xm,(m*n,1))
Ymr=np.reshape(Ym,(m*n,1))
Xproof=np.append(Xmr,Ymr,axis=1)
Yg=reg_log(B,Xproof)
Z=np.reshape(Yg,(m,n))
Z=np.round(Z)
plt.figure(figsize=(10,10))
plt.contour(Xm,Ym,Z)
plt.scatter(X[:,0],X[:,1],c=Y,edgecolors='w')
plt.show()
```
